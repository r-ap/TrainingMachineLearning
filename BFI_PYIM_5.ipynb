## Menggabungkan DataFrame

### **pd.Concat**

```python
city_data = pd.DataFrame(
    {"revenue":[4200,6500,8000], "employee_count":[5,8,3]},index=['Amsterdam','Toronto','Tokyo'])
city_data
```

```python
further_city_data = pd.DataFrame(
    {"revenue": [7000, 3400], "employee_count": [2, 2]},
    index=["New York", "Barcelona"])
further_city_data
```

```python
all_city_data = pd.concat([city_data, further_city_data])
all_city_data
```

```python
city_countries = pd.DataFrame({
    "country": ["Holland", "Japan", "Holland", "Canada", "Spain"],
    "capital": [1, 1, 0, 0, 0]},
    index=["Amsterdam", "Tokyo", "Rotterdam", "Toronto", "Barcelona"]
)
city_countries
```

```python
pd.concat([all_city_data,city_countries])
```

```python
pd.concat([all_city_data,city_countries],axis=1)
```

### **pd.Merge**

Menggabungkan dua dataframe berdasarkan satu kolom sebagai kunci

`pd.merge(data1,data2, on='', left_on='', left_index='', right_on='',right_on='',how='')`


```python
countries = pd.DataFrame({
    "population_millions": [17, 127, 37],
    "continent": ["Europe", "Asia", "North America"]
}, index=["Holland", "Japan", "Canada"])
countries
```

```python
city_countries
```

```python
pd.merge(city_countries,countries, left_on='country',right_index = True, how='inner')
```

```python
pd.merge(city_countries,countries, left_on='country',right_index = True, how='left')
```
