## WordCloud

```python
!pip install -q wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
!wget --quiet https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/alice_novel.txt -O alice_novel.txt

text = open('alice_novel.txt','r').read()
```

```python
stopwords = set(STOPWORDS)

stopwords
```

```python
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
alice_wc.generate(text)
```

```python
import matplotlib.pyplot as plt

plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```

```python
stopwords.add('said') # add the words said to stopwords

# re-generate the word cloud
alice_wc.generate(text)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```

```python
import pandas as pd

e_com = pd.read_csv('https://raw.githubusercontent.com/fahmimnalfrzki/Dataset/main/sales_data_sample.csv',encoding='Latin-1')
e_com.head()
```

```python
e_com['COUNTRY'].unique()
```

```python
e_com['CITY'].unique()
```

```python
country_text = ' '.join(e_com['COUNTRY'].values)
print(country_text)
```

```python
country_wc = WordCloud(
    background_color='white',
    max_words=2000)

country_wc.generate(country_text)

fig,ax = plt.subplots(figsize=(14,18))
ax.imshow(country_wc)
ax.axis('off')
plt.show()
```

```python
city_text = ' '.join(e_com['CITY'].values)

city_wc = WordCloud(
    background_color='white',
    max_words=100)

city_wc.generate(city_text)

fig,ax = plt.subplots(figsize=(14,18))
ax.imshow(city_wc)
ax.axis('off')
plt.show()
```

```python
cities = []

for x in e_com['CITY']:
  if len(x.split(' '))==1:
    cities.append(x)
  else:
    cities.append('-'.join(x.split(' ')))

cities = ' '.join(cities)
print(cities)
```

```python
city_wc = WordCloud(
    background_color='white',
    max_words=100)

city_wc.generate(cities)

fig,ax = plt.subplots(figsize=(14,18))
ax.imshow(city_wc)
ax.axis('off')
plt.show()
```

### Map with Data

```python
df_incidents = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Police_Department_Incidents_-_Previous_Year__2016_.csv')
df_incidents.head()
```

```python
limit = 100
df_incidents = df_incidents.iloc[0:limit, :]
```

```python
# San Francisco latitude and longitude values
latitude = 37.77
longitude = -122.42
```

```python
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# display the map of San Francisco
sanfran_map
```

```python
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

sanfran_map
```

```python
# add incidents to map
sanfran_map.add_child(incidents)
```

```python
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add pop-up text to each marker on the map
latitudes = list(df_incidents.Y)
longitudes = list(df_incidents.X)
labels = list(df_incidents.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sanfran_map)    
    
# add incidents to map
sanfran_map.add_child(incidents)
```

```python
# create map and display it
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# loop through the 100 crimes and add each to the map
for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.CircleMarker(
        [lat, lng],
        radius=5, # define how big you want the circle markers to be
        color='yellow',
        fill=True,
        popup=label,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(sanfran_map)

# show map
sanfran_map
```

```python
from folium import plugins

# let's start again with a clean copy of the map of San Francisco
sanfran_map = folium.Map(location = [latitude, longitude], zoom_start = 12)

# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(sanfran_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

# display map
sanfran_map
```
